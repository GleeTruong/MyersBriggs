```python
import pandas as pd
import nltk
import re
import numpy as np
import os
import string
import emot
from emot.emo_unicode import UNICODE_EMO, EMOTICONS #These are Dict
```

```python
#Allows us to see the whole string
pd.set_option('display.max_colwidth', -1)
```

```python
rawFile = pd.read_csv('mbti_1.csv')
rawFile["posts"]= rawFile['posts'].astype(str)
#rawFile.index = [np.arange(len(rawFile.index)), rawFile.index] #Adding an index to the df
```

```python
def cleaner(comments):
    #comments = re.sub('http\S+ ','',comments)#removing web addresses
    comments = re.sub('\w+:\/\/\S+','',comments)#removing web addresses
    comments = re.sub('\.\.\.','', comments)#removing triple...
    comments = re.sub('[0-9]+','', comments)#removing numbers
    comments = re.sub('\|\|\|', '', comments)# removing delimiters
    for emote, mean in EMOTICONS.items():
        comments = comments.replace(emote, ' <'+mean+'> ')
    comments = re.sub(r':(\w+):', r' <\1> ', comments) # we are capturing group1 :group1: and returning <group1>
    #comments = comments.split('|||') We can use this if we wan t a list of delimiters
    #comments = ''.join([word.lower() for word in comments if word not in string.punctuation])#Removing punctuations
    return comments
rawFile['Clean']=rawFile['posts'].apply(lambda x: cleaner(x))
```

```python
df = rawFile
df.to_csv("Clean_v1.csv")
```
